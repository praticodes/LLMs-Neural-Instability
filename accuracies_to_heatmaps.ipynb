## Labile State vs Non-Labile State Age Analysis

### Labile State Age Results

```python
# Read the dataframes for the various saved percentages of labuile state (instability active) age training testing on BBQ
labile_100 = pd.read_csv("/content/drive/MyDrive/1025_instability_age_eval_results_100steps")
labile_200 = pd.read_csv("/content/drive/MyDrive/1025_instability_age_eval_results_200steps")
labile_15p = pd.read_csv("/content/drive/MyDrive/high_instability_age_eval_results_15p")
labile_33p = pd.read_csv("/content/drive/MyDrive/high_instability_age_eval_results_33p")
labile_100p = pd.read_csv("/content/drive/MyDrive/high_instability_age_eval_results_100p")
```

```python
# Compute accuracy for each
labile_100_acc = labile_100["correct"].mean() * 100
labile_200_acc = labile_200["correct"].mean() * 100
labile_15p_acc = labile_15p["correct"].mean() * 100
labile_33p_acc = labile_33p["correct"].mean() * 100
labile_100p_acc = labile_100p["correct"].mean() * 100

# Report said accuracy
print(labile_100_acc, labile_200_acc, labile_15p_acc, labile_33p_acc, labile_100p_acc)
```

### Non-Labile State Age Results

```python
# Read the dataframes for the various saved percentages of non-labile state (instablility deactivated) age training testing on BBQ
non_labile_100 = pd.read_csv("/content/drive/MyDrive/zero_instability_age_eval_results_100steps")
non_labile_200 = pd.read_csv("/content/drive/MyDrive/zero_instability_age_eval_results_200steps")
non_labile_15p = pd.read_csv("/content/drive/MyDrive/zero_instability_age_eval_results_15p")
non_labile_33p = pd.read_csv("/content/drive/MyDrive/zero_instability_age_eval_results_900steps")
non_labile_50p = pd.read_csv("/content/drive/MyDrive/zero_instability_age_eval_results_50p")
non_labile_100p = pd.read_csv("/content/drive/MyDrive/zero_instability_age_eval_results_fullsteps")
```

```python
# Compute accuracy for each
non_labile_100_acc = non_labile_100["correct"].mean() * 100
non_labile_200_acc = non_labile_200["correct"].mean() * 100
non_labile_15p_acc = non_labile_15p["correct"].mean() * 100
non_labile_33p_acc = non_labile_33p["correct"].mean() * 100
non_labile_50p_acc = non_labile_50p["correct"].mean() * 100
non_labile_100p_acc = non_labile_100p["correct"].mean() * 100

# Report said accuracy
print(non_labile_100_acc, non_labile_200_acc, non_labile_15p_acc, non_labile_33p_acc, non_labile_50p_acc, non_labile_100p_acc)
```

### Visuals Generation

```python
# Heatmap comparing accuracies

heatmap_data = {
    '100 steps': [non_labile_100_acc, labile_100_acc],
    '200 steps': [non_labile_200_acc, labile_200_acc],
    '15%': [non_labile_15p_acc, labile_15p_acc],
    '33%': [non_labile_33p_acc, labile_33p_acc],
    '100%': [non_labile_100p_acc, labile_100p_acc],
}

df = pd.DataFrame(heatmap_data, index=['Non-Labile', 'Labile'])

plt.figure(figsize=(10, 5))

# Heatmap from Red (for 0%) to Green (for 100%), 2f accuracy displays
acc_cmap = plt.get_cmap('RdYlGn')
acc_cmap.set_bad('lightgrey')
sns.heatmap(df, annot=True, cbar_kws={'label': 'Accuracy', 'location': 'left'}, cmap=acc_cmap, vmin=0, vmax=100, fmt='.2f')

plt.title("LLM Accuracy on BBQ: Labile State Induction vs Non-Labile")
plt.show()
```
